```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("GroceryStoreDataSet.csv", names =['Products'])
df.head()
```

```python
df.shape
```

```python
data = list(df["Products"].apply(lambda x:x.split(',')))
data 
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
freq_items = apriori(df, min_support = 0.2, use_colnames = True, verbose = 1) 
```

```python
freq_items
```

```python
association_rules(freq_items, metric="confidence", min_threshold = 0.6)
```

### (Milk -> Bread) 
- Milk ve bread tüm alışverişlerin yüzde 20'singe görülmektedir.
- Milk alan müşterilerin yüzde 80'si bread de almaktadır.
- Milk alınan alışverişlerde bread ürününün satışı 1.23 kat artmaktadır.
