## Kurs Performansına göre yeni yetenek keşfi

Kursun 4 haftalık süreci bitmiş durumda. Bu 4 haftalık süreçte kursiyerlerin derslere ve mentor görüşmelerine katılımı, zorunlu ve seçmeli ödevlerin kaçını yaptığı, udemy deki derslerde nerede olduklarının kaydı tutulmuştur. Mentorler öğrencilerine teknik ve uyum puanı vermişlerdir.

4 haftalık süreçte:
- 8 Ders, 3 Mentor Görüşmesi yapıldı.
- 8 Zorunlu, 9 Seçmeli ödev verildi.
- Udemydeki 15 dersin bitirilmesi istendi.





### Veri Seti

- id - her bir öğrenci için unique
- Öğrenci_Adı_Soyadı - öğrencinin ismi ve soyadı
- Mentor_Adı - öğrencinin mentorunun adı
- Katıldığı_Ders - öğrencinin katıldığı toplam ders sayısı
- Katıldığı_Mentor_Görüşmesi - öğrencinin katıldığı top mentor görüşmesi sayısı
- Zorunlu_Ödev - öğrencinin yaptığı toplam zorunlu ödev sayısı
- Seçmeli_Ödev - öğrencinin yaptığı toplam seçmeli ödev sayısı
- Udemy_İzlenilen_Ders - öğrencinin udemyde izledi toplam bölüm sayısı
- Grup_Numarası - öğrencinin bulunduğu grubun numarası
- Mentor_Teknik_Puanı - mentorun öğrenciye verdiği teknik puan (1-5)
- Mentor_Uyum_Puanı - mentorun öğrenciye verdiği uyum puanı (1-5)
- Durum - öğrencinin üniversite, yüksek lisans veya çalışıyor olduğunu gösteriyor

# RFM ANALİZİ

Recency : en son ne zaman mentor görüşmesine katıldı

Frequency : toplam kaç der ve mentor görüşmesine katılmıştır


```python
df.head()
```

```python
recency_df.rename(columns={"En_Son_Katılım":"Recency"}, inplace = True)
```

```python
recency_df.head()
```

```python
freq_df = df[["Öğrenci_Adı_Soyadı","Toplam_Katılım"]]
```

```python
freq_df.rename(columns={"Toplam_Katılım":"Frequency"}, inplace = True)
```

```python
rfm = df[["Öğrenci_Adı_Soyadı","En_Son_Katılım","Toplam_Katılım"]]
```

```python
rfm.rename(columns={"En_Son_Katılım":"Recency"}, inplace = True)
rfm.rename(columns={"Toplam_Katılım":"Frequency"}, inplace = True)
```

```python
rfm.head()
```

```python
rfm["FrequencyScore"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
rfm["RecencyScore"] = rfm["Recency"]
```

```python
rfm.head()
```

```python
(rfm["RecencyScore"].astype(str)+
rfm["FrequencyScore"].astype(str)).head()
```

```python
# rfm skorları kategorik değere dönüştürülüp df'e eklendi
rfm["RFM_SCORE"] = (rfm['RecencyScore'].astype(str) + 
                    rfm['FrequencyScore'].astype(str))
```

```python
rfm.head()
```

```python
rfm.describe().T
```

```python
# Regular Expressions (Düzenli İfadeler) kullanılarak RFM haritası çıkarıldı
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
```

```python
rfm.head()
```

```python
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
```

```python
rfm.head()
```

```python
rfm[["Segment","Recency","Frequency"]].groupby("Segment").agg(["mean","median","count"])
```
